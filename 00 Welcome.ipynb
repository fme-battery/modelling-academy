### Welcome to FME Battery academy 2026

This repository contains course material for the battery testing data processing tutorial.

**Authors**: Jan Petter Mæhlen, Julia Wind, Inger-Emma Nylund, Meghana Sudarshan, Williams Agyei Appiah, Svenas Burba, Lorena Hendrix, Xavier Raynaud  
**Location**: IFE, Norway  
**Date**: 2026.05.28  

## Agenda for today

- 13:00 - 13:15 Welcome and checking in to the FME Battery Modelling Academy Hub
- 13:15 - 13:45 Introduction to `cellpy` with practical sessions and exercise
- 13:45 - 14:00 Using `cellpy` for generating `ds-Toolbox`-friendly files
- 14:00 - 14:15 Break
- 14:15 - 15:00 Introduction to `ds-Toolbox` and extraction of entropy
- 15:00 - 15:05 Break
- 15:05 - 15:30 Introduction to `BatMo`
- 15:30 - 16:00 Simulating with and without entropy contribution with `BatMo`
