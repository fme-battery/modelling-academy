### 3. Experimental data conversion

In this session we are going to convert the experimental data from excel to a more python evironment friendly .pkl format.

First we import packages:

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

We begin by calling in cell IDs and temperatures at which the data was collected in form of list variables.

```python
cellIDs = ["BattMo"]      # Identifiers for each cell
order = [45, 35, 25, 15]  # Temperatures for each (°C)
```

Then we inform python what headers are relevant and need to be kept, at the same time the converted file will only keep the headers and respective data in header columns.

```python
headers = [
    "Step Index / 1",
    "Voltage / V",
    "Charge Capacity / Ah",
    "Discharge Capacity / Ah"
]

# Corresponding field names to use internally
fields = ["step", "voltage", "chgAh", "disAh"]
column_map = dict(zip(headers, fields))

# Field names for the four different testing scripts
stepFields = ["script1", "script2", "script3", "script4"]
```

Now we can loop through the data with defined indentity information and convert the desired datasets from excel to .pkl Python data structure with following code.

```python
for cellID in cellIDs:

    dirname = cellID.split("_")[0]  # remove "_" and everything after

    for temp in order:

        # Build file prefix
        if temp < 0:
            prefix = f"DATA/{dirname}_OCV/{cellID}_OCV_N{abs(temp):02d}"
        else:
            prefix = f"DATA/{dirname}_OCV/{cellID}_OCV_P{temp:02d}"

        data = {}

        for script in [1, 3]:

            file_path = Path(f"{prefix}_S{script}.csv") #file path, name and format

            if not file_path.exists():
                print(f"File not found: {file_path}")
                continue

            print(f"Reading {file_path}")

            #------------------- Read Excel Files ------------------
            #data = pd.ExcelFile(file_path)
            #OCVData = []

            #for sheet in data.sheet_names:
                #if sheet == "Info":
                    #continue

                #print(f"Processing sheet {sheet}")
                #df = pd.read_excel(file_path, sheet_name=sheet)

                #df_filtered = df[headers].rename(columns=column_map)
                #OCVData.append(df_filtered)
            #-------------------------------------------------------

            #------------------- Read CSV Files --------------------
            OCVData = []

            print(f"Processing {file_path}")
            df = pd.read_csv(file_path)

            df_filtered = df[headers].rename(columns=column_map)
            OCVData.append(df_filtered)
            #-------------------------------------------------------
            
            if OCVData:
                combined_df = pd.concat(OCVData, ignore_index=True)

                # 🔹 Save directly next to original file
                out_file = file_path.with_suffix(".pkl")
                combined_df.to_pickle(out_file)

                print(f"Saved {out_file}")
print ("Files processed.")
```
